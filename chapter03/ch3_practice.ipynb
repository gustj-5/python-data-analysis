## 3.1.2 조건문
조건에 따라 데이터 분류

```python
temperature = 25

if temperature > 30:
    print("날씨가 매우 덥습니다!")
elif temperature >20:
    print("날씨가 따듯합니다!")
else:
    print("날씨가 쌀쌀합니다!")
```

## 3.1.3 반복문
여러 데이터에 동일한 연산 적용

```python
scores = [73, 83, 92]

for i in scores:
    print(i + 5)
```

## 3.1.4 함수
매개변수로 값을 받아 결과 반환

```python
def avg(values):
    return sum(values) / len(values)

avg(scores)
```
