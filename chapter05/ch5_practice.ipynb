## 5.1.1 CSV 파일 불러오기

```python
import pandas as pd

df = pd.read_csv('data/example.csv') # CSV 파일 불러오기
print(df) # 데이터 출력
```

## 5.1.2 엑셀 파일 불러오기

```python
df = pd.read_excel('data/example.xlsx') # 엑셀 파일 불러오기
print(df) # 데이터 출력
```

## 5.1.3 다양한 옵션
- 데이터 파일에서 특정 열만 불러오기
- 열 이름(헤더)가 없는 파일 불러오기

```python
# 특정 열만 불러오기
df = pd.read_csv('data/example.csv', usecols=['Name', 'Age'])
print(df)
```

```python
# 헤더 없는 파일 불러오기
df = pd.read_csv('data/example_no_header.csv', header=None)
print(df)
```

## 5.2.3 통계 요약하기

```python
# 기본 통계 정보 확인
print(df.describe())
```

```python
# 비수치형 데이터도 포함한 통계 정보 확인
print(df.describe(include='all'))
```

## 5.3.1 평균 구하기

```python
# Age 열의 평균
print(df['Age'].mean())
```

```python
# Salary 열의 평균
print(df['Salary'].mean())
```

```python
# 여러 열의 평균
print(df[['Age', 'Salary']].mean())
```

## 5.3.2 최댓값 구하기

```python
print(df.max())
```

```python
# Salary 열의 최댓값
print(df['Salary'].max())
```

## 5.3.3 최솟값 구하기

```python
print(df.min())
```

```python
# Age 열의 최솟값
print(df['Age'].min())
```

## 5.3.4 데이터 더하기

```python
print(df.sum())
```

```python
# 수치형 데이터만 더하기
print(df.sum(numeric_only=True))
```
