## 4.1 Pandas 사용하기

```python
import pandas as pd
```

## 4.2 시리즈 다루기
index 매개변수를 사용해 인덱스 지정 가능

```python
# 시리즈 생성
fruits = ["apple", "banana", "pineapple", "cherry"]
idx = ["a", "b", "c", "d"]
series_fruits = pd.Series(fruits, index=idx)

print(series_fruits) 
print(series_fruits["b"]) # 인덱스를 통해 데이터 접근
```

## 4.3.1 데이터프레임 다루기
행과 열 인덱스 지정 가능

```python
# 데이터프레임 생성
data = [
    ["Alice", 25, "USA"],
    ["Jenny", 28, "Korea"],
    ["Mark", 26, "Canada"]
]
df = pd.DataFrame(data, index=["Person1", "Person2", "Person3"], columns=["Name", "Age", "Country"])

print(df)

print("단일 열 접근")
print(df["Name"])

print("여러 열 접근")
print(df[["Name", "Country"]])
```

## 4.3.2 loc / iloc
loc와 iloc를 통해 데이터프레임의 데이터 접근하기

```python
# Person1 행의 Name과 Age 데이터 접근
print(df.loc["Person1", ["Name", "Age"]])
```

```python
# 두 번째 행의 첫 번째와 세 번째 열 데이터 접근
print(df.iloc[1, [0, 2]])
```
